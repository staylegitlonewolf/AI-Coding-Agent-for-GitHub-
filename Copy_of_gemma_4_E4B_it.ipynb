```python
# Load model directly
from transformers import AutoProcessor, AutoModelForImageTextToText

processor = AutoProcessor.from_pretrained("google/gemma-4-E4B-it")
model = AutoModelForImageTextToText.from_pretrained("google/gemma-4-E4B-it")
```

### Demonstrate model usage

Let's try to use the model with a sample image and text prompt. We'll use `Pillow` to handle the image and `requests` to fetch it from a URL.

```python
from PIL import Image
import requests

# Sample image URL
image_url = "https://picsum.photos/200/300" # Changed to a reliably publicly accessible image

# Fetch the image
response = requests.get(image_url, stream=True)
response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
image = Image.open(response.raw).convert("RGB")

print("Image loaded successfully.")
# You can display the image if you like:
# display(image)
```

```python
# Prepare the input for the model
text_prompt = "What is in the image?"
inputs = processor(text=text_prompt, images=image, return_tensors="pt")

print("Inputs prepared successfully.")
```

```python
# Generate text using the model
generated_ids = model.generate(**inputs, max_new_tokens=100)

# Decode the generated tokens
generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("Generated Text:")
print(generated_text)
```
